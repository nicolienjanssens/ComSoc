```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("results.csv")
```

```python
df.keys()
```

```python
x = range(1, 10)
means = []
div = []

for i in x:
    means.append(df[f"k_{i}_util"].mean())
    div.append(df[f"k_{i}_util"].std())

plt.bar(x, means, yerr=div)

plt.title("utility", fontsize=20)
plt.xticks(range(1,10))

plt.xlabel("k-value", fontsize=15)
plt.ylabel("score", fontsize=15)
plt.show()
```
